## Run full set of models 
For each model we sweep through a set of parameter values and store the result for each setting. Results are written to a pickle file in ../output/results. For example, model_code_13_F10.pickle stores results for the code_13 model using the F10 measure. If the result file for a given model already exists in ../output/results, then that file is not overwritten and the model is not run again.

```python
runmodel(modelnames = ['model_n', 'model_n_nodilation', 'model_n_nodistance', 'model_knn', 'model_knn_thresh', 'model_singleton', 'model_onebig'], optimize=2, showgraph=0)
```

There are 16 variants of the CODE model. For each one we first run a gridsearch that sweeps through a set of parameter values:

```python
codemodelnames = ['model_code_' + str(i) for i in np.arange(15)]
runmodel(modelnames = codemodelnames, optimize=2, showgraph=0)
```

For each variant we then run an optimization using Powell's conjugate direction method initialized using the best value found in the grid search. Results are written to a separate pickle file : for example, model_code_13_F10_opt.pickle stores the results for the optimization for the code_13 model.

```python
runmodel(modelnames = codemodelnames, optimize=1, showgraph=0)
```
